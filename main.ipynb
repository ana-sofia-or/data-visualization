```python
import pandas as pd
from ydata_profiling import ProfileReport
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sleep_health_df = pd.read_csv('sleep_health_and_lifestyle.csv')
profile = ProfileReport(sleep_health_df, title="Sleep Health Profiling Report")
profile.to_file("sleep_health_and_lifestyle.html")
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=sleep_health_df, x='Gender', y='Sleep Duration', hue='Stress Level', palette='pastel')
plt.title('Gender vs Sleep Duration Grouped by Stress Level')
plt.xlabel('Gender')
plt.ylabel('Sleep Duration')
plt.legend(title='Stress Level')
plt.grid(True)
plt.show()

#Com este gráfico de barras é possível perceber que os homens são ligeramente menos afetados pelos níveis de stress comparando com as mulheres. Conseguimos perceber em ambos os géneros que o nível de stress está relacionado com a quantidade de sono, pelo que quanto mais stress a pessoa tiver menos horas de sono conseguirá ter.
```

```python
plt.figure(figsize=(10, 6))

sns.kdeplot(data=sleep_health_df, x='Sleep Duration', hue='Quality of Sleep', fill=True, palette='viridis', alpha=0.5)
plt.title('Kernel Density Estimate of Sleep Duration by Quality of Sleep')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Density')
plt.show()

#Com este gráfico de densidade é possível perceber que a qualidade de sono está relacionada com a quantidade de sono. Pelo que as pessoas que dormem mais possuem a qualidade de sono mais elevada. Nestes dados conseguimos perceber também que existe uma maioria de pessoas a dormir perto de 6 horas, tendo uma qualidade de sono razoável.
```

```python
plt.figure(figsize=(12, 6))

sns.histplot(data=sleep_health_df, x='Age', hue='Sleep Disorder', multiple='stack', palette='viridis')

plt.title('Histogram of Age by Sleep Disorder')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()

#Com este gráfico (histograma) conseguimos perceber que entre os 20 e 40 anos existe uma quantidade pequena de pessoas com insónias ou apneia do sono. Conseguimos também concluir que a maioria das pessoas entre 42 e 45 anos têm uma destas duas disturbios de sono. Apesar disto, verifica-se uma distribuição anormal dos disturbios de sono em relação à idade, pelo que não se pode concluir a 100% que existe correlação.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sleep_health_df, x='Physical Activity Level', y='Quality of Sleep', color='red', hue='Daily Steps')
plt.title('Physical Activity Level vs Quality of Sleep')
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.grid(True)
plt.show()

#Através deste gráfico de dispersão conseguimos identificar que as pessoas com mais passos diários têm uma melhor qualidade de sono. Conseguimos também perceber que quanto menor for o nível de atividade física pior será a qualidade de sono.
```
